```python
import pandas as pd
import numpy as np
import seaborn as sns
import math
import matplotlib.pyplot as plt

# plt.rcParams['backend'] = "qt4agg"
# plt.rcParams['backend.qt4'] = "PySide"

# наименование длительности курса
name_sz = {'sz_code': [0, -10, -20, -30, -40], \
           'sz_name': ["?", "Экспресс-курс", "Ограниченный", "Полноценный", "Длительный"]}
name_sz = pd.DataFrame(data=name_sz)
name_sz['sz_code'] = name_sz['sz_code'].astype('int')

# наименование цели курса
name_aimi = {'aimi_code': [-30, -20, -10, 0, 10, 20, 30, 40], \
           'aimi_name': ["Для себя", "Для жизни", "Для работы", "?", "Повыш. усп. втор.", "Втор. экз., олимп.", "Повыш. усп.", "Важный экзамен"]}
name_aimi = pd.DataFrame(data=name_aimi)
name_aimi['aimi_code'] = name_aimi['aimi_code'].astype('int')

data = pd.read_csv('express_mine(-20).tsv', delimiter = '\t' \
            , dtype = {'order_typ_new': 'str',\
                       'order_typ_old': 'str'
                       })
# new_raw_pricing_data.tsv

# data = pd.read_csv('all_orders_3_year.tsv', delimiter = '\t' \
#             , dtype = {'order_typ_new': 'str',\
#                        'order_typ_old': 'str'
#                        })


data['new_group_mismatch'] = np.where(data['order_typ_old'] != data['order_typ_new'], 'mismatch', '+')

data_mismatch = data[data['new_group_mismatch'] == 'mismatch']
data_mismatch = data_mismatch[data_mismatch['order_typ_old'].notnull()]
data_mismatch['aimi'] = data_mismatch['aimi'].astype('int')
data_mismatch['sz'] = data_mismatch['sz'].astype('int')
data = data[data['order_typ_old'].notnull()]
data_mismatch = pd.merge(data_mismatch, name_sz, left_on = 'sz', right_on = 'sz_code', how = 'left')
data_mismatch = pd.merge(data_mismatch, name_aimi, left_on = 'aimi', right_on = 'aimi_code', how = 'left')


all_ds_orders = data_mismatch['order_id'].count()
ds_to_all = data_mismatch['order_id'].count() / data['order_id'].count()
change_in_express = (data_mismatch['order_id'].count() + \
                    data[data['order_typ_old'].str.contains('Экспресс')]['order_id'].count()) \
                     / data[data['order_typ_old'].str.contains('Экспресс')]['order_id'].count() - 1


print("Всего заказов поменяли категорию: {a} \
       \n Итого изменили группу: {b} \
       \n Увеличили экспресс группу на : {c} \
       ".format(a=all_ds_orders, b=ds_to_all, c=change_in_express))
```

```python
# group categories to analyse
data_mismatch_group_by_type = data_mismatch.groupby('order_typ_old', as_index = False)['order_id'].count()
data_mismatch_group_by_newtype = data_mismatch.groupby('order_typ_new', as_index = False)['order_id'].count()
data_mismatch_group_by_month = data_mismatch.groupby('month_receivd', as_index = False)['order_id'].count()
data_mismatch_group_by_aim = data_mismatch.groupby('aimi_name', as_index = False)['order_id'].count()
data_mismatch_country = data_mismatch.groupby('country', as_index = False)['order_id'].count()
data_mismatch_sz = data_mismatch.groupby('sz_name', as_index = False)['order_id'].count()
```

```python
plt.figure(figsize=(5,6))
sns.set_style("darkgrid")
plot_1 = sns.barplot(x='order_typ_old', y='order_id', data=data_mismatch_group_by_type)
plot_1.set_xticklabels(plot_1.get_xticklabels(), rotation=90)
plt.savefig('check12.png', bbox_inches = 'tight')
```

```python
fig, axes = plt.subplots(ncols=2, nrows=2, figsize=(19,22))
# sns.set_style("darkgrid")
for ax in fig.axes:
    plt.sca(ax)
    plt.xticks(rotation=90)
    plt.tick_params(axis='both', which='major', labelsize=11)
# q = sns.barplot(x='order_typ_old', y='order_id', data=data_mismatch_group_by_type, ax = axes[0][0])
# axes[0][0].set_xticklabels(q.get_xticklabels(), rotation=90)
sns.barplot(x='month_receivd', y='order_id', data=data_mismatch_group_by_month, palette = 'Set2', ax = axes[0][0])
sns.barplot(x='aimi_name', y='order_id', data=data_mismatch_group_by_aim, palette = 'Set2', ax = axes[0][1])
sns.barplot(x='country', y='order_id', data=data_mismatch_country, palette = 'Set2', ax = axes[1][0])
sns.barplot(x='sz_name', y='order_id', data=data_mismatch_sz, palette = 'Set2', ax = axes[1][1])
plt.savefig('check1.png', bbox_inches = 'tight')
```

```python
plt.clf()
plt.figure(figsize=(6,5))
plot_7 = sns.barplot(x='order_typ_new', y='order_id', data=data_mismatch_group_by_newtype)
plot_7.set_xticklabels(plot_7.get_xticklabels(), rotation=90)
plt.savefig('data_mismatch_group_by_newtype1.pdf')
```
